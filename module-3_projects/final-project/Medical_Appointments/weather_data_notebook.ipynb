```python
weatherStation= 'ws-836490'
searchdate = pd.Timestamp('2016-06-15')
searchdate.to_datetime
searchdate.month
#https://www.tutiempo.net/clima/05-2016/ws-836490.html
url = 'https://www.tutiempo.net/clima/'+ str(searchdate.month).zfill(2)  + '-'+ str(searchdate.year) +'/'+ weatherStation +'.html'  
session = requests.session()
resp = session.get(url)

```

```python
#isplay(HTML(resp.text))
```

```python
soup = BeautifulSoup(resp.text, 'html.parser')
```

```python
soup.title
```

```python
raw_table = soup.find('table', {'class':'medias mensuales'})
```

```python
rows = []
for row in raw_table.children:
    rows.append(row.get_text(separator= ','))
rows
```

```python
table = [row.split(',') for row in rows if len(row) > 1]
```

```python
print(*table, sep = '\n')
```
